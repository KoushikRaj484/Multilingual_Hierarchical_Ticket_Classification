## Clean special characters, HTML tags, and extra symbols, and lowercasing Tokenization, stopword removal.

```python
import re
import pandas as pd
import nltk

from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from unidecode import unidecode

nltk.download('punkt')
nltk.download('stopwords')
nltk.download('punkt_tab')
stop_english = set(stopwords.words('english'))
stop_german = set(stopwords.words('german'))

def clean_text(t):
    if pd.isna(t):
        return ''
    
    # Convert accented characters (√© ‚Üí e)
    # t = unidecode(str(t))
    # ‚úÖ Convert to lowercase for NLP
    t = t.lower()

    tokens = word_tokenize(t)  
    # tokenize ['This', 'is', 'an', 'example',
    # 'sentence', ',', 'demonstrating', 'word', 'tokenization', 'with', 'NLTK', '!']
    tokens = [i for i in tokens if (i not in stop_english and i not in stop_german) and len(i) > 2]  # remove stopwords + short words
    t = ' '.join(tokens)
       
    # Remove HTML tags
    t = re.sub(r'<.*?>', ' ', t)

    t = re.sub(r'\\n', ' ', t)
    # Remove URLs
    t = re.sub(r'http\S+|www\.\S+', ' ', t)
    
    # Remove email addresses
    t = re.sub(r'\S+@\S+', ' ', t)
    
    # üî• Remove all special characters, punctuation, underscores, hyphens, and escape sequences
    t = re.sub(r'[%\[\]_\\<\(\]#\?\'\":\)\-\;\+\!\/,>\.\n\r]', ' ', t)
    
    # Remove multiple spaces and trim
    t = re.sub(r'\s+', ' ', t).strip()
    
    return t
    
# Apply to DataFrame
df['cleaned_text'] = df['text'].apply(clean_text)
```

## Encode targets:
+ type, queue: LabelEncoder + One-hot encoding.
+ tags: MultiLabelBinarizer.

```python
# Optional: Encode 'priority' and 'language' for future model enhancements
from sklearn.preprocessing import LabelEncoder, OneHotEncoder, MultiLabelBinarizer
from tensorflow.keras.utils import to_categorical

le_priority = LabelEncoder()
le_language = LabelEncoder()

df['priority_encoded'] = le_priority.fit_transform(df['priority'])
df['language_encoded'] = le_language.fit_transform(df['language'])


# Encode 'type' and 'queue'
le_type = LabelEncoder()
le_queue = LabelEncoder()

df['type_encoded'] = le_type.fit_transform(df['type'])
df['queue_encoded'] = le_queue.fit_transform(df['queue'])

ohe = OneHotEncoder(sparse_output=False)
type_ohe = ohe.fit_transform(df[['type_encoded']])
queue_ohe = ohe.fit_transform(df[['queue_encoded']])


# --- Prepare targets ---
# Type (categorical)
y_type_int = df["type_encoded"].values
y_type = to_categorical(y_type_int)


# Queue (categorical)
y_queue_int = df["queue_encoded"].values
y_queue = to_categorical(y_queue_int)

# Tags (multi-label)
tags_cols = ['tag_1','tag_2','tag_3','tag_4','tag_5','tag_6','tag_7']
df['tags_combined'] = df[tags_cols].astype(str).values.tolist()
# clean tags: remove 'nan', 'None', 'Unknown' variations if present
def clean_tags_list(lst):
    out = []
    for t in lst:
        if t is None:
            continue
        ts = str(t).strip()
        if ts.lower() in ['', 'nan', 'none', 'unknown', 'nan.0']:
            continue
        out.append(ts)
    return out

df['tags_combined'] = df['tags_combined'].apply(clean_tags_list)
mlb = MultiLabelBinarizer()
y_tags = mlb.fit_transform(df['tags_combined'])
```

```python
import numpy as np
from gensim.models import Word2Vec
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences

MAX_NUM_WORDS = 20000
MAX_SEQ_LEN = 200
EMBEDDING_DIM = 300
# LSTM_UNITS = 128
# DROPOUT_RATE = 0.3
BATCH_SIZE = 64
EPOCHS = 10


texts = df['cleaned_text'].tolist() 
tokenizer = Tokenizer(num_words=20000, oov_token="<OOV>") 
tokenizer.fit_on_texts(texts) 
sequences = tokenizer.texts_to_sequences(texts) 
word_index = tokenizer.word_index 
num_words = min(MAX_NUM_WORDS, len(word_index) + 1)
max_len = max(len(seq) for seq in sequences)

X = pad_sequences(sequences, maxlen=len(max(sequences,key = len)), padding='pre', truncating='pre')
X = np.array(X, dtype="int32")
# Tokenize text for Word2Vec
sentence = [text.split() for text in df['cleaned_text']]

# Train Word2Vec model
w2v_model = Word2Vec(sentence, vector_size=300, window=5, min_count=2, workers=4)

embedding_matrix = np.zeros((num_words, 300))
for word, i in word_index.items():
    if i >= num_words:
        continue
    if word in w2v_model.wv:
        embedding_matrix[i] = w2v_model.wv[word]
    # else:
    #     # small random init for OOV within vocab range
    #     embedding_matrix[i] = np.random.normal(scale=0.01, size=(300,))

```

```python
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader, TensorDataset
import optuna
from optuna.samplers import GridSampler

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("Using device:", device)
X_train_t = torch.tensor(X_train, dtype=torch.long).to(device)

y_type_t  = torch.tensor(y_type_train, dtype=torch.float32).to(device)
y_queue_t = torch.tensor(y_queue_train, dtype=torch.float32).to(device)
y_tags_t  = torch.tensor(y_tags_train, dtype=torch.float32).to(device)

train_ds = TensorDataset(X_train_t, y_type_t, y_queue_t, y_tags_t)
train_loader = DataLoader(train_ds, batch_size=900, shuffle=True)


class MultiTaskModel(nn.Module):
    def __init__(self, num_words, embedding_matrix, nu, seq_len,
                 type_dim, queue_dim, tags_dim):
        super(MultiTaskModel, self).__init__()

        # --- Embedding ---
        self.embedding = nn.Embedding(num_words, 300)
        self.embedding.weight = nn.Parameter(
            torch.tensor(embedding_matrix, dtype=torch.float32)
        )
        self.embedding.weight.requires_grad = False  # non-trainable

        # --- Shared BiLSTM Encoder ---
        self.lstm = nn.LSTM(
            input_size=300,
            hidden_size=nu,
            batch_first=True,
            bidirectional=True
        )

        # --- Heads ---
        self.type_head = nn.Sequential(
            nn.Linear(nu * 2, nu),
            nn.ReLU(),
            nn.Linear(nu, type_dim)
        )

        self.queue_head = nn.Sequential(
            nn.Linear(nu * 2 + type_dim, nu),
            nn.ReLU(),
            nn.Linear(nu, queue_dim)
        )

        self.tags_head = nn.Sequential(
            nn.Linear(nu * 2 + queue_dim, nu),
            nn.ReLU(),
            nn.Linear(nu, tags_dim)
        )

    def forward(self, x):
        x = self.embedding(x)

        _, (h, _) = self.lstm(x)
        h = torch.cat((h[0], h[1]), dim=1)   # BiLSTM concat (2*nu)

        type_out = self.type_head(h)
        queue_in = torch.cat([h, type_out], dim=1)
        queue_out = self.queue_head(queue_in)

        tags_in = torch.cat([h, queue_out], dim=1)
        tags_out = self.tags_head(tags_in)

        return type_out, queue_out, tags_out
```

```python
def train_model(model, optimizer, criterion1, criterion2, criterion3):
    model.train()
    total_loss = 0

    for X, y1, y2, y3 in train_loader:
        optimizer.zero_grad()

        o1, o2, o3 = model(X)

        loss1 = criterion1(o1, y1)
        loss2 = criterion2(o2, y2)
        loss3 = criterion3(o3, y3)

        loss = loss1 + loss2 + loss3
        loss.backward()
        optimizer.step()

        total_loss += loss.item()

    return total_loss / len(train_loader)
```

```python
def objective(trial):
    nu = trial.suggest_categorical("nurons", [32, 64, 128])

    model = MultiTaskModel(
        num_words=num_words,
        embedding_matrix=embedding_matrix,
        nu=nu,
        seq_len=X_train.shape[1],
        type_dim=y_type.shape[1],
        queue_dim=y_queue.shape[1],
        tags_dim=y_tags.shape[1]
    ).to(device)

    optimizer = optim.Adam(model.parameters(), lr=1e-3)

    criterion1 = nn.CrossEntropyLoss()
    criterion2 = nn.CrossEntropyLoss()
    criterion3 = nn.BCEWithLogitsLoss()

    val_losses = []

    for epoch in range(10):
        loss = train_model(model, optimizer, criterion1, criterion2, criterion3)
        val_losses.append(loss)

    return val_losses[-1]
```
